```python
import random
# Function to generate different DNA sequences based on barcodes
def generate_sequences(barcodes, length):
    sequences = []
    for n in range(3):
      for barcode in barcodes.values():
        new_seq = ""
        for i in range(length):
            x = random.randrange(1,100)
            nucleotide = barcode[x % len(barcode)]
            new_seq += str(nucleotide)

        sequences.append(Seq(new_seq))
    return sequences
```

#Needleman-Wunsch Global Alignment

```python
def Identify_Species_Global(sequence, barcodes):
    MaxScore = 0
    Scores = []
    identified_species = None
    for species, barcode in barcodes.items():
        alignment = align.globalxx(sequence, barcode, one_alignment_only=True)[0]
        score = alignment.score
        Scores.append(score)

        if score > MaxScore:
            MaxScore = score
            identified_species = species
    return identified_species , MaxScore , Scores
```

#Alignment with affine gap penalties

```python
def HighestScoringAlignment(v, w, ScoreMatrix, GOP , GEP):
    n = len(v)
    m = len(w)

    #Initialize Score Matrix
    S = []
    for i in range(n+1):
      row = []
      for j in range(m+1):
         row.append(0)
      S.append(row)

    #Initialize Horizontal Gap Matrix
    Horizontal = []
    for i in range(n+1):
        row = []
        for j in range(m+1):
            row.append(float('-inf'))
        Horizontal.append(row)

    #Initialize Vertical Gap Matrix
    Vertical = []
    for i in range(n+1):
        row = []
        for j in range(m+1):
            row.append(float('-inf'))
        Vertical.append(row)

    TracebackMatrix = []
    for i in range(n+1):
        row = []
        for j in range(m+1):
            row.append('-')
        TracebackMatrix.append(row)


    # Initialize first row of both Horizontal Gap Matrix and Score Matrix
    for i in range(1, n+1):
        Horizontal[i][0] = GOP + (i-1) * GEP
        S[i][0] = Horizontal[i][0]

    # Initialize first column of both Vertical Gap Matrix and Score Matrix
    for j in range(1, m+1):
        Vertical[0][j] = GOP + (j-1) * GEP
        S[0][j] = Vertical[0][j]

    # Fill in the matrices
    for i in range(1, n+1):
        for j in range(1, m+1):
            S[i][j] = max(S[i-1][j-1] + ScoreMatrix[v[i-1]][w[j-1]],
                          Horizontal[i-1][j-1] + ScoreMatrix[v[i-1]][w[j-1]],
                          Vertical[i-1][j-1] + ScoreMatrix[v[i-1]][w[j-1]])

            Horizontal[i][j] = max(S[i-1][j] + GOP + GEP,
                                   Horizontal[i-1][j] + GEP)

            Vertical[i][j] = max(S[i][j-1] + GOP + GEP,
                                 Vertical[i][j-1] + GEP)

            if S[i][j] == max(S[i][j],Horizontal[i][j],Vertical[i][j]):
                TracebackMatrix[i][j] = 'D' #This is for Diagonal traceback
            elif Horizontal[i][j] == max(S[i][j],Horizontal[i][j],Vertical[i][j]):
                TracebackMatrix[i][j] = 'U' #This is for Up traceback
            else:
                TracebackMatrix[i][j] = 'L' #This is for Left traceback

    # Traceback
    V_aligned = []
    W_aligned = []

    i, j = n, m
    while i > 0 or j > 0:
        if i > 0 and j > 0 and TracebackMatrix[i][j] == 'D':
            V_aligned.append(v[i-1])
            W_aligned.append(w[j-1])
            i -= 1
            j -= 1
        elif i > 0 and TracebackMatrix[i][j] == 'U':
            V_aligned.append(v[i-1])
            W_aligned.append('-')
            i -= 1
        elif j > 0 and TracebackMatrix[i][j] == 'L':
            V_aligned.append('-')
            W_aligned.append(w[j-1])
            j -= 1
        elif i>0 and TracebackMatrix[i][j] == '-':
            V_aligned.append(v[i-1])
            W_aligned.append('-')
            i -= 1
        elif j>0 and TracebackMatrix[i][j] == '-':
            V_aligned.append('-')
            W_aligned.append(w[j-1])
            j -= 1
        else:
            break

    V_aligned = ''.join(reversed(V_aligned))
    W_aligned = ''.join(reversed(W_aligned))

    return V_aligned, W_aligned, S[n][m] , S , Horizontal , Vertical , TracebackMatrix
```

```python
print("Using Alignment with affine gap penalties")

ScoreMatrix = {
    'A': {'A': 1, 'C': -1, 'G': -1, 'T': -1},
    'C': {'A': -1, 'C': 1, 'G': -1, 'T': -1},
    'G': {'A': -1, 'C': -1, 'G': 1, 'T': -1},
    'T': {'A': -1, 'C': -1, 'G': -1, 'T': 1}
}

GOP = -2
GEP = -1

V_aligned, W_aligned, Score , S , Horiz , Ver , TM = HighestScoringAlignment( generated_sequences[0],barcodes['Dolphin'], ScoreMatrix, GOP, GEP)
print("Sequence 1 Alignment :", V_aligned)
print("Sequence 2 Alignment :", W_aligned)
print("Alignment Score:", Score)
```
